```python
yelp_review= pd.read_csv('yelp_restraunt_reviews.csv')
yelp_review.head(10)
```

```python
yelp_users=pd.read_csv('yelp_restraunt_raters.csv')
yelp_users.head(10)
```

```python
top_users = 15
user_most_reviews = yelp_review.groupby(['user_id']).size().reset_index(name='counts')
top_users_ids = user_most_reviews.sort_values(['counts'])[::-1][:top_users].user_id.values
d = {'Top users id': top_users_ids}
topUserDf = pd.DataFrame(d)
topUserDf
```

```python
#Extracting friends and fans of these top users
user_frnds = {}
for  users_id in top_users_ids:
    
    record = yelp_users.loc[yelp_users['user_id'] == users_id]
    user_frnds[users_id] = {}
    user_frnds[users_id]["user_name"]= record.user_name.values[0]
    user_frnds[users_id]["friends"]= record.friends.values[0]
    if record.friends is not "None":
        user_frnds[users_id]["friends_count"]= len(record.friends.values[0].split(","))
    else:
        user_frnds[users_id]["friends_count"]=0
    user_frnds[users_id]["user_review_count"]=record.user_review_count.values[0]
    user_frnds[users_id]["n_fans"]=record.n_fans.values[0]
```

```python
#Extraction results
user_friend_df = pd.DataFrame.from_dict(data= user_frnds,orient="index")
user_friend_df
```

```python
#Visualizing the results obtained
axes = user_friend_df.plot(x="user_name", y=["user_review_count", "friends_count", "n_fans"],
                           kind="bar",figsize=(12, 7))
plt.xlabel('Name', axes=axes)
plt.ylabel('Count', axes=axes)
plt.title('Top User Review trend', axes=axes)
plt.savefig('topUserTrend', bbox_inches='tight')
plt.show()
```

```python
business_rank = 10 # 0 to 9
temp = yelp_review[["business_id","rating"]]
five_star_reviews = temp[temp["rating"]>4]
trending_business_reviews = five_star_reviews.groupby(["business_id"]).size().reset_index(name='counts')
business_id = trending_business_reviews.sort_values(['counts'])[::-1][:10].business_id.values[:business_rank]
business_name = []
for i in business_id:
    j = df.loc[df['business_id'] == i].business_name.values[0]
    business_name.append(j)
d = {'Buisness_Id':business_id,'Buisness_Name':business_name}
dataframe = pd.DataFrame(d)
dataframe
```

```python
business_rank = 0
temp = yelp_review[["business_id","rating"]]
five_star_reviews = temp[temp["rating"]>4]
trending_business_reviews = five_star_reviews.groupby(["business_id"]).size().reset_index(name='counts')

business_id = trending_business_reviews.sort_values(['counts'])[::-1][:10].business_id.values[business_rank]
business_name = df.loc[df['business_id'] == business_id].business_name.values[0]
```

```python
user_reviws = yelp_review.loc[yelp_review['business_id']==business_id]
topuser_reviws = user_reviws.groupby(['user_id']).size().reset_index(name='counts')
print("Total users who gave ratings to {} are {}".format(business_name,len(topuser_reviws)))
```

```python
top_users = 1530
topuser_reviws = topuser_reviws.sort_values(['counts'])[::-1][:top_users]
```

```python
users_ids = topuser_reviws.user_id.values
users_ids = pd.merge(topuser_reviws,right=yelp_users, how="inner",on='user_id')
users_ids =users_ids [["user_name","user_id", "friends"]]
```

```python
def calc_frnd_review(frnds, business_id, yelp_review):
        frnds = frnds.split(',')
        friend_review = yelp_review.loc[(yelp_review['business_id']==business_id) &
                                        (yelp_review['user_id'].isin(frnds))
                                       ][["rating","user_id"]]
        friend_review_cnt = len(friend_review.user_id.values)
        friends_ratings = []
        if(friend_review_cnt>0):
            friends_ratings.append(friend_review.rating)
            total_stars = friend_review.rating.sum()
        else:
            total_stars = 0
        return  friend_review_cnt,total_stars,friends_ratings

users_ids[["frnd_count","total_rating", "friends_ratings"]]= users_ids["friends"].apply(lambda frnds: pd.Series(calc_frnd_review(frnds, business_id, yelp_review), index=['frnd_count','total_rating', 'friends_ratings']))
users_ids[users_ids["frnd_count"]>0]
```

```python
axes = users_ids[users_ids["frnd_count"]==1].plot(x="user_name", y=["total_rating"],
                           kind="bar",figsize=(20, 17))
plt.xlabel('Name', axes=axes)
plt.ylabel('Rating', axes=axes)
plt.title('Ratings of Pai Northern Thai Kitchen by top users friends'.format(business_name), axes=axes)
plt.show()
(len(users_ids[users_ids["frnd_count"]==1]))
```
