```
users = pd.read_csv('yelp_restraunt_raters.csv')
reviews = pd.read_csv('yelp_restraunt_reviews.csv')
restraunts = pd.read_csv('yelp_restraunts.csv')
```

```
restrauntlist = restraunts['categories'].tolist()
```

```
import collections as c

uniquewords = c.Counter()
for i in restrauntlist:
  words = i.split(', ')
  uniquewords.update(words)


del uniquewords['Restaurants']
del uniquewords['Food']

top100 = uniquewords.most_common(100)
print(set(top100))
print(top100[1][0])
print(type(top100))
categorylist = []
categorycounts = []
for i in top100:
  categorylist.append(i[0])
  categorycounts.append(i[1])
```

```
restfinal=[]
countfinal=[]
ratingfinal=[]

for i in categorylist:
  count = 0
  rating = 0
  final = 0
  for index, row in restraunts.iterrows():
    if i in row['categories']:
      count += 1
      rating += row['stars']
      final = rating / count
  if final > 0:
    restfinal.append(i)
    countfinal.append(count)
    ratingfinal.append(final)
   
```

```
d = {'Category': restfinal, 'Count': countfinal, 'Rating Average': ratingfinal}
df = pd.DataFrame(data=d)
df.sort_values(by=['Rating Average'], ascending=False)
```

```
df.to_csv('top100ratings.csv')
files.download('top100ratings.csv')
```
